```python
dem_txn=pd.read_csv('Demand_txn_updated.csv')
```

#### a. Handle the missing values in the column Percent_Quantity_With_Discount with strategy = mean 

```python
dem_txn[dem_txn.isnull().any(axis=1)]['Percent_Quantity_With_Discount']
```

```python
dem_txn.Percent_Quantity_With_Discount.mean()
```

```python
dem_txn['Percent_Quantity_With_Discount'] = dem_txn['Percent_Quantity_With_Discount'].fillna(dem_txn.Percent_Quantity_With_Discount.mean())
```

```python
dem_txn['Percent_Quantity_With_Discount']
```

#### b.	One hot encode the columns Mapped_Sales_Type 

```python
dem_txn['Mapped_Sales_Type'].value_counts()
```

```python
dem_txn1 = pd.get_dummies(dem_txn, columns=['Mapped_Sales_Type'],prefix=['Type'])
```

```python
dem_txn1[['Type_Bakery','Type_Delivery','Type_Dine-in','Type_Express']]
```

#### c.	Label binarize the column Day

```python
from sklearn.preprocessing import LabelBinarizer
```

```python
lb_style = LabelBinarizer()
```

```python
lb_result = lb_style.fit_transform(dem_txn1['Day'])
```

```python
lb_result
```

```python
Day_conv = pd.DataFrame(lb_result, columns=lb_style.classes_)
```

```python
lb_style.classes_
```

```python
pd.concat([dem_txn1,Day_conv],axis=1)
```
