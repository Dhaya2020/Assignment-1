#### 13.	Perform following aggregation task on below table
a.	Get the average age group by Regrion
b.	Get the sum of Age and Validation group by Hair color and then reset the index
c.	Get the man and the median of Age group by Region


```python
df5=pd.DataFrame([[48,'Black','SW',1],[22,'Black','SW',0],[27,'Brown','NE',0],[28,'Black','SE',3],[21,'Blond','SE',1], [51,'Black','NW',0], [23,'Brown','SE',1]],columns =['AGE','HAIR COLOR','US REGION','VALIDATION'])
```

```python
df5
```

```python
#a.	Get the average age group by Regrion
df5.groupby(['US REGION'])['AGE'].sum()
```

```python
df5.groupby(['HAIR COLOR'])['AGE','VALIDATION'].sum()
```

```python
# b.	Get the sum of Age and Validation group by Hair color and then reset the index
df5.groupby(['HAIR COLOR'])['AGE','VALIDATION'].sum().reset_index()
```

```python
#c.	Get the man and the median of Age group by Region
df5.groupby(['US REGION'])['AGE'].agg([np.mean, np.median]).reset_index()
```
